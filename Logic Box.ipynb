```python
print("Welcome to the Pattern Generator and Number Analyzer!")

while True:
    print()
    print("Select an option:")
    print("1. Generate a Pattern")
    print("2. Analyze a Range of Numbers")
    print("3. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        rows = int(input("Enter the number of rows for the pattern: "))

        if rows > 0:
            print()
            print("Pattern:")
            for i in range(1, rows + 1):
                print("*" * i)
        else:
            print("Invalid input! Rows must be positive.")

    elif choice == 2:
        print()

        start = int(input("Enter the start of the range: "))
        end = int(input("Enter the end of the range: "))

        if end >= start:
            total = 0
            for num in range(start, end + 1):
                if num % 2 == 0:
                    print("Number " + str(num) + " is Even")
                else:
                    print("Number " + str(num) + " is Odd")
                total = total + num

            print("Sum of all numbers from " + str(start) + " to " + str(end) + " is: " + str(total))
        else:
            print("Invalid range! End must be greater than or equal to start.")

    elif choice == 3:
        print("Exiting the program. Goodbye!")
        break

    else:
        print("Invalid choice! Please try again.")
```
